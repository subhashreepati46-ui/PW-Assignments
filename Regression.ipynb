Question 1:  What is Simple Linear Regression?

Simple Linear Regression Definition
Simple linear regression is a statistical method that models the relationship between two continuous quantitative variables using a straight line. One variable serves as the independent or predictor variable (typically denoted as x), while the other is the dependent or response variable (denoted as y). The model estimates how changes in the predictor variable affect the response variable.​

Model Equation

The relationship follows the equation y^= β0 + β1x where y^ is the predicted value of the dependent variable, β0 is the y-intercept, and β1 is the slope indicating the change in y per unit change in x. This line minimizes the sum of squared differences between observed and predicted values, known as ordinary least squares.

Key Assumptions
The model assumes linearity between variables, independence of observations, homoscedasticity (constant error variance), and normality of residuals. Violations can lead to unreliable predictions or inferences.

Question 2: What are the key assumptions of Simple Linear Regression?

Simple linear regression relies on a set of conditions about the relationship between variables and the model errors. These assumptions ensure that estimation, hypothesis tests, and predictions are valid.

Main assumptions

1 Linearity

The relationship between the independent variable x and the dependent variable y is assumed to be linear, meaning the true mean of y changes in a straight-line fashion with x
In practice, this is often checked using a scatter plot of x vs y
to see whether the points roughly follow a straight-line pattern.

2 Independence of observations (and residuals)

Each observation in the dataset should be independent of the others, so that the errors (residuals) are not correlated with each other.​
This is especially important in time series or clustered data, where dependence can appear as patterns in residual plots over time or groups.

3 Homoscedasticity (constant variance of errors)

The variance of the error terms should be constant across all values of the predictor variable, a property called homoscedasticity.​
When this assumption holds, residuals show roughly equal spread around the regression line at all levels of
x; if spread increases or decreases systematically, heteroscedasticity is present.

4 Normality of errors

The error terms (or equivalently, the residuals) are assumed to follow a normal distribution with mean zero. Normality supports the use of t-tests and confidence intervals; it is commonly checked using histograms or Q–Q plots of residuals.

5 No major outliers or influential points


Simple linear regression assumes there are no extreme outliers or highly influential observations that can unduly distort the fitted line.​
Such points are usually diagnosed via residual plots and influence measures (for example, leverage or Cook’s distance) and may require separate treatment or robust methods.

Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

Heteroscedasticity refers to a situation in regression where the variance of the error terms (residuals) is not constant across all levels of the independent variable. In other words, the spread of residuals changes (for example, increases or decreases) as the predicted values or the predictor variable change, instead of staying roughly the same.

Why it is a problem


*   Ordinary Least Squares (OLS) regression assumes homoscedasticity, meaning the errors all have the same variance; heteroscedasticity violates this key assumption.

*   When heteroscedasticity is present, the coefficient estimates can still be unbiased, but their standard errors become unreliable, which affects t-tests, confidence intervals, and F-tests, leading to incorrect conclusions about which predictors are statistically significant.

Practical impact on regression models


*   Under heteroscedasticity, hypothesis tests may indicate that variables are significant when they are not (or miss significance when it actually exists), because standard errors are underestimated or overestimated.

*  This reduces the efficiency and reliability of the model’s inferences and forecasts, so detecting and correcting heteroscedasticity (for example, by transforming variables, using weighted least squares, or using robust standard errors) is important for valid regression analysis.   






Question 4: What is Multiple Linear Regression?

Multiple linear regression extends simple linear regression to model the relationship between one dependent variable and two or more independent variables. It uses a linear equation to predict the dependent variable based on the combined effects of multiple predictors.

Model Equation

The general form is y^=β0 +β1 x1 +β2 x2 +⋯+βp xp, where y ^is the predicted value, β0 is the intercept, and each βi represents the coefficient for independent variable xi  . Coefficients are estimated via ordinary least squares to minimize the sum of squared residuals.

Key Differences from Simple Regression



*   Handles multiple predictors simultaneously, allowing assessment of each variable's impact while controlling for others.

*   Assumes linearity, independence, homoscedasticity, normality of residuals, and no perfect multicollinearity among predictors.






Question 5: What is polynomial regression, and how does it differ from linear
regression?

Polynomial regression models nonlinear relationships between variables using polynomial functions of the predictor(s), such as quadratic or cubic terms. It fits a curve rather than a straight line to capture patterns like U-shapes or S-curves in data.​

Model Form

The equation takes the form y = β0 + β1x + β2x2 + ⋯ + βmxm +ϵ y=β 0 where higher powers of x (e.g., x2  ) allow curvature. Despite the nonlinearity in x, it remains a linear model in the parameters β, solved via least squares.

Differences from Linear Regression


*   Relationship type: Linear regression assumes a straight-line (first-degree) fit; polynomial handles curves by adding higher-degree terms.​

*  Flexibility: Polynomial captures complex patterns linear models miss but risks overfitting with high degrees, requiring regularization.​


*  Assumptions: Shares linear regression assumptions (e.g., independence, homoscedasticity) but checks for curvature via scatterplots first.






Question 6:  Implement a Python program to fit a Simple Linear Regression model to
the following sample data:


*   X = [1, 2, 3, 4, 5]
  
*   Y = [2.1, 4.3, 6.1, 7.9, 10.2]


Plot the regression line over the data points.

(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Get predictions and coefficients
Y_pred = model.predict(X)
slope = model.coef_[0]
intercept = model.intercept_

print("Simple Linear Regression Results:")
print(f"Slope (β1): {slope:.4f}")
print(f"Intercept (β0): {intercept:.4f}")
print(f"Equation: y = {intercept:.4f} + {slope:.4f}x")
print("\nActual vs Predicted Values:")
for i in range(len(X)):
    print(f"X={X[i][0]:.0f}: Actual={Y[i]:.1f}, Predicted={Y_pred[i]:.2f}")

# Plot the data points and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', s=100, label='Data points')
plt.plot(X, Y_pred, color='red', linewidth=2, label=f'Regression line: y = {intercept:.2f} + {slope:.2f}x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression: Data Points and Fitted Line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Question 7: Fit a Multiple Linear Regression model on this sample data:

· Area = [1200, 1500, 1800, 2000]

· Rooms = [2, 3, 3, 4]

· Price = [250000, 300000, 320000, 370000]

Check for multicollinearity using VIF and report the results.

(Include your Python code and output in the code box below.)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import warnings
warnings.filterwarnings('ignore')

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

print("Dataset:")
print(df)
print("\n" + "="*60)

# Fit Multiple Linear Regression model
X = df[['Area', 'Rooms']]
X_const = add_constant(X)  # Add intercept for VIF
y = df['Price']

model = LinearRegression()
model.fit(X, y)

# Predictions and model statistics
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

print("Multiple Linear Regression Results:")
print(f"Coefficients: Area = {model.coef_[0]:.2f}, Rooms = {model.coef_[1]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"R² Score: {r2:.4f}")
print(f"Equation: Price = {model.intercept_:.0f} + {model.coef_[0]:.2f}×Area + {model.coef_[1]:.2f}×Rooms")
print("\nActual vs Predicted Prices:")
for i in range(len(df)):
    print(f"Area={df['Area'][i]}, Rooms={df['Rooms'][i]}: Actual=${y[i]:,}, Predicted=${y_pred[i]:,.0f}")

print("\n" + "="*60)

# Check for multicollinearity using VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

print("Variance Inflation Factor (VIF) Results:")
print(vif_data)
print("\nVIF Interpretation:")
print("- VIF < 5: No multicollinearity")
print("- VIF 5-10: Moderate multicollinearity")
print("- VIF > 10: High multicollinearity")

# Correlation matrix
print("\nCorrelation Matrix:")
print(df[['Area', 'Rooms']].corr())

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Actual vs Predicted
axes[0].scatter(y, y_pred, color='blue', s=100)
axes[0].plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
axes[0].set_xlabel('Actual Price')
axes[0].set_ylabel('Predicted Price')
axes[0].set_title('Actual vs Predicted Prices')

# 2. 3D scatter plot with regression plane
from mpl_toolkits.mplot3d import Axes3D
ax = fig.add_subplot(133, projection='3d')
ax.scatter(df['Area'], df['Rooms'], df['Price'], color='blue', s=100)
# Create meshgrid for regression plane
xx, yy = np.meshgrid(np.linspace(df['Area'].min(), df['Area'].max(), 10),
                     np.linspace(df['Rooms'].min(), df['Rooms'].max(), 10))
zz = model.intercept_ + model.coef_[0]*xx + model.coef_[1]*yy
ax.plot_surface(xx, yy, zz, alpha=0.5, color='red')
ax.set_xlabel('Area')
ax.set_ylabel('Rooms')
ax.set_zlabel('Price')
ax.set_title('3D Regression Plane')

plt.tight_layout()
plt.show()
```

Question 8: Implement polynomial regression on the following data:

· X = [1, 2, 3, 4, 5]

· Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.

(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pandas as pd

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

print("Dataset:")
df = pd.DataFrame({'X': X.flatten(), 'Y': Y})
print(df)
print("\n" + "="*60)

# Fit 2nd-degree Polynomial Regression
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

# Predictions
Y_pred = model.predict(X_poly)
r2 = r2_score(Y, Y_pred)

# Get polynomial coefficients
coef1 = model.coef_[0]  # x term
coef2 = model.coef_[1]  # x^2 term
intercept = model.intercept_

print("Polynomial Regression Results (2nd Degree):")
print(f"Equation: y = {intercept:.3f} + {coef1:.3f}x + {coef2:.3f}x²")
print(f"R² Score: {r2:.4f}")
print("\nActual vs Predicted Values:")
for i in range(len(X)):
    print(f"X={X[i][0]:.0f}: Actual={Y[i]:.1f}, Predicted={Y_pred[i]:.2f}")

print("\n" + "="*60)

# Plot the data points and polynomial curve
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', s=100, label='Data points', zorder=5)

# Generate smooth curve for plotting
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
Y_plot = model.predict(X_plot_poly)

plt.plot(X_plot, Y_plot, color='red', linewidth=3,
         label=f'2nd Degree Polynomial: y = {intercept:.2f} + {coef1:.2f}x + {coef2:.2f}x²')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (2nd Degree) Fit')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Compare with linear regression
linear_model = LinearRegression()
linear_model.fit(X, Y)
Y_linear = linear_model.predict(X_plot)
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', s=100, label='Data points', zorder=5)
plt.plot(X_plot, Y_plot, color='red', linewidth=3,
         label=f'Polynomial (R²={r2:.3f}): y = {intercept:.2f} + {coef1:.2f}x + {coef2:.2f}x²')
plt.plot(X_plot, Y_linear, color='green', linewidth=2, linestyle='--',
         label=f'Linear (R²={linear_model.score(X,Y):.3f})')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial vs Linear Regression Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:

· X = [10, 20, 30, 40, 50]

· Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.

(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from scipy import stats

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

print("Dataset:")
df = pd.DataFrame({'X': X.flatten(), 'Y': Y})
print(df)
print("\n" + "="*50)

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

print("Simple Linear Regression Results:")
print(f"Slope: {model.coef_[0]:.3f}")
print(f"Intercept: {model.intercept_:.3f}")
print(f"R² Score: {r2_score(Y, Y_pred):.4f}")

print("\nData Summary:")
summary_df = pd.DataFrame({
    'X': X.flatten(),
    'Y_Actual': Y,
    'Y_Predicted': np.round(Y_pred, 2),
    'Residuals': np.round(residuals, 3)
})
print(summary_df)
print("\n" + "="*50)

# Residuals Plot for Heteroscedasticity Assessment
plt.figure(figsize=(12, 5))

# Plot 1: Residuals vs Fitted Values (Main heteroscedasticity check)
plt.subplot(1, 2, 1)
plt.scatter(Y_pred, residuals, color='blue', s=120, alpha=0.7, edgecolors='black')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Fitted Values (Predicted Y)')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values\n(Homoscedasticity Check)')
plt.grid(True, alpha=0.3)

# Plot 2: Residuals vs X
plt.subplot(1, 2, 2)
plt.scatter(X, residuals, color='green', s=120, alpha=0.7, edgecolors='black')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('X Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predictor (X)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Breusch-Pagan test for formal heteroscedasticity assessment
X_sm = sm.add_constant(X)
model_sm = sm.OLS(Y, X_sm).fit()
bp_test = sm.stats.diagnostic.het_breuschpagan(model_sm.resid, model_sm.model.exog)

print("Heteroscedasticity Assessment:")
print("\nBreusch-Pagan Test Results:")
print(f"LM Statistic: {bp_test[0]:.4f}")
print(f"P-value: {bp_test[1]:.4f}")

print("\nVisual Assessment from Residuals Plot:")
print("- No systematic pattern in residuals vs fitted values")
print("- Spread of residuals appears roughly constant across fitted values")
print("- Points randomly scattered around zero line")

print("\nConclusion:")
if bp_test[1] > 0.05:
    print("✓ Homoscedasticity assumption holds (p = {:.4f} > 0.05)".format(bp_test[1]))
else:
    print("✗ Evidence of heteroscedasticity (p ≤ 0.05)")
```

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

To address heteroscedasticity and multicollinearity in a real estate house price prediction model, follow a systematic diagnostic and remediation process.

Diagnose Issues First

Heteroscedasticity Detection:



*   Create residuals vs fitted values plot to identify funnel patterns

*   Run Breusch-Pagan or White test (p < 0.05 indicates heteroscedasticity)

*   Check scale-location plot (√|residuals| vs fitted)


Multicollinearity Detection:

*   Calculate Variance Inflation Factors (VIF > 5-10 problematic)
*   Review correlation matrix between predictors
*   Examine condition number (>30 indicates issues)



Address Heteroscedasticity










```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Sample data for demonstration (re-using data from Question 7, adjusting column names)
data = {
    'area': [1200, 1500, 1800, 2000],
    'rooms': [2, 3, 3, 4],
    'price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Define X and y for the OLS model
X = df[['area', 'rooms']]
y = df['price']
# Add constant for statsmodels OLS/WLS
X_const = sm.add_constant(X)

# Step 1: Transform dependent variable
df['log_price'] = np.log(df['price'])  # Log transformation

# Step 2: Weighted Least Squares if transformation insufficient
# For WLS, weights are typically inversely proportional to the variance of the error terms.
# Here, it's given as 1 / df['area'] which is a common heuristic if variance increases with area.
weights = 1 / df['area']  # Weight by inverse variance proxy
model_wls = sm.WLS(y, X_const, weights=weights).fit() # Use X_const here for model fitting

# Step 3: Robust standard errors
model_robust = sm.OLS(y, X_const).fit(cov_type='HC3') # Use X_const here for model fitting

print("Log-transformed prices added to df:")
print(df)
print("\nWLS Model Summary:")
print(model_wls.summary())
print("\nOLS Model with Robust Standard Errors Summary:")
print(model_robust.summary())
```

Common fixes:

*  Log-transform price (most effective for real estate)

*  Use robust standard errors (HC3 or HC1)

*  Weighted least squares with area-based weights

*  Generalized least squares if variance structure known


Address Multicollinearity


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import RidgeCV
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Define sample data
data = {
    'area': [1200, 1500, 1800, 2000],
    'rooms': [2, 3, 3, 4],
    'price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n" + "="*60 + "\n")

# 2. Create predictor variable DataFrame X and response variable Series y
X = df[['area', 'rooms']]
y = df['price']

# 3. Add a constant term to X to create X_const for statsmodels models
X_const = sm.add_constant(X)

# 4. Apply a log transformation to the price variable
df['log_price'] = np.log(df['price'])
y_log = df['log_price']

print("DataFrame with log_price:")
print(df)
print("\n" + "="*60 + "\n")

# 5. Fit a Weighted Least Squares (WLS) model
# Here, weights are typically inversely proportional to the variance of the error terms.
# Using 1 / df['area'] as a proxy, assuming variance increases with area.
weights = 1 / df['area']
model_wls = sm.WLS(y, X_const, weights=weights).fit()
print("WLS Model Summary (Original Price):")
print(model_wls.summary())
print("\n" + "="*60 + "\n")

# 6. Fit an Ordinary Least Squares (OLS) model with robust standard errors (original y)
model_ols_robust_orig_y = sm.OLS(y, X_const).fit(cov_type='HC3')
print("OLS Model with HC3 Robust Standard Errors (Original Price):")
print(model_ols_robust_orig_y.summary())
print("\n" + "="*60 + "\n")

# 7. Calculate Variance Inflation Factors (VIF) for all features in X_const
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]
print("Variance Inflation Factor (VIF) Results for X_const:")
print(vif_data)
print("\nVIF Interpretation:")
print("- VIF < 5: No multicollinearity")
print("- VIF 5-10: Moderate multicollinearity")
print("- VIF > 10: High multicollinearity")
print("\n" + "="*60 + "\n")

# 8. Apply Ridge Regression to X (without the constant) and y_log
# We use RidgeCV to find the best alpha (regularization strength)
ridge_model = RidgeCV(alphas=np.logspace(-6, 6, 13))
ridge_model.fit(X, y_log)
print("Ridge Regression Model (on log_price):")
print(f"Best Alpha: {ridge_model.alpha_:.6f}")
print(f"Intercept: {ridge_model.intercept_:.4f}")
print(f"Coefficients: {ridge_model.coef_}")
print(f"  Area coefficient: {ridge_model.coef_[0]:.4f}")
print(f"  Rooms coefficient: {ridge_model.coef_[1]:.4f}")
print("\n" + "="*60 + "\n")

# 9. Fit a final OLS model with y_log as the dependent variable and X_const as independent variables,
#    using 'HC3' robust standard errors.
final_ols_model_robust_log_y = sm.OLS(y_log, X_const).fit(cov_type='HC3')
print("Final OLS Model with HC3 Robust Standard Errors (Log-transformed Price):")
print(final_ols_model_robust_log_y.summary())
print("\n" + "="*60 + "\n")
```

Strategies:

* Remove highly correlated features (e.g., drop "total_rooms" if keeping "bedrooms" + "bathrooms")

* Principal Component Analysis (PCA) for dimensionality reduction

* Ridge regression (L2 penalty shrinks correlated coefficients)

* Feature engineering (combine area + location into price_per_sqm)

Build Robust Final Model

```python
# 1. Feature selection based on VIF < 5 (Note: Original features 'area' and 'rooms' have high VIF, so using them as is for demonstration.)
X_clean = X[['area', 'rooms']] # Use existing features from X

# 2. Log transformation for heteroscedasticity
y_log = np.log(df['price'])

# 3. Ridge regression for remaining collinearity
from sklearn.linear_model import RidgeCV
ridge = RidgeCV(alphas=np.logspace(-6, 6, 13))
ridge.fit(X_clean, y_log)

# 4. Final model with robust errors for inference
final_model = sm.OLS(y_log, sm.add_constant(X_clean)).fit(cov_type='HC3')
print(final_model.summary())
```
