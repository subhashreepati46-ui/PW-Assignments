1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?


Type I and Type II errors are two types of errors that can occur in hypothesis testing, impacting decision-making by leading to incorrect conclusions about the population being studied.

Type I error, also known as a false positive, occurs when the null hypothesis (which states there is no effect or difference) is rejected even though it is actually true. This error means concluding that there is an effect or difference when there isn't one in reality. For example, diagnosing a healthy person as having a disease when they do not actually have it is a Type I error. The probability of making a Type I error is denoted by the significance level alpha (α).

Type II error, also known as a false negative, happens when the null hypothesis is not rejected even though it is false. This error means failing to detect a real effect or difference. For example, concluding that a drug is not effective when it actually is would be a Type II error. The probability of making a Type II error is denoted by beta (β), and the power of the test (1 - β) is the probability of correctly rejecting a false null hypothesis.

The impact on decision-making is significant because:


*   A Type I error can lead to adopting a false claim or treatment, resulting in wasted resources or harm.

*   A Type II error can lead to missed opportunities to identify real effects or benefits, resulting in inaction.

There is a trade-off between these errors: lowering the risk of a Type I error increases the risk of a Type II error, and vice versa. Researchers must balance these risks by setting appropriate significance levels and designing studies with sufficient power.

2 : What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

The P-value in hypothesis testing is a measure of how likely it is to observe your data, or something more extreme, if the null hypothesis is true. It quantifies the probability of your results happening by random chance under the assumption that there is no real effect or difference.

When interpreting the P-value:


*   A small P-value (commonly less than 0.05) suggests strong evidence against the null hypothesis, so you may reject the null hypothesis in favor of the alternative hypothesis.

*  A large P-value (greater than 0.05) indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis; the results are consistent with what would be expected by random chance.

The P-value does not measure the probability that the null hypothesis is true; rather, it tells you how compatible your data are with the assumption of the null hypothesis. Researchers set a significance level (alpha, often 0.05) before testing to decide at what threshold the P-value leads to rejecting or failing to reject the null hypothesis. Ultimately, the P-value guides decision-making by indicating whether an observed effect is statistically significant or likely due to random variation

3 : Explain the difference between a Z-test and a T-test, including when to use
each.

A Z-test and a T-test are both statistical tests used to compare sample data and draw conclusions about a population, but they differ mainly in their assumptions about sample size and knowledge of population variance.

A Z-test is used when the sample size is large (generally greater than 30) and the population variance is known. It relies on the normal distribution, making it suitable for situations where data is plentiful and population parameters are well established. The Z-test is commonly applied for comparing population means and proportions in large samples.​

A T-test, on the other hand, is used when the sample size is small (less than 30) and the population variance is unknown. In this case, the test relies on the t-distribution, which accounts for extra uncertainty by having heavier tails. The T-test is ideal for situations with limited data and when you must estimate the population standard deviation using your sample.​

To summarize:



*   Use a Z-test if you have a large sample and know the population variance.

*   Use a T-test if you have a small sample or do not know the population variance.

Choosing the correct test ensures accurate results in hypothesis testing, and helps in making appropriate statistical decisions.

4 : What is a confidence interval, and how does the margin of error influence
its width and interpretation?

A confidence interval is a range of values, calculated from sample data, that likely contains the true value of an unknown population parameter (such as the mean) with a specified level of confidence—commonly 95% or 99%. Instead of giving just a single estimate, the confidence interval provides upper and lower bounds which express the uncertainty inherent in sampling.​

The margin of error is the amount added and subtracted from the sample statistic to create the interval. It directly influences the width of the confidence interval: a larger margin of error means a wider interval, showing greater uncertainty, while a smaller margin of error produces a narrower interval, showing higher precision. The margin of error depends on sample size, data variability, and the chosen confidence level—a higher confidence level or more variable data increases the margin of error and thus widens the interval.​

Interpreting a confidence interval means that, for example with 95% confidence, if the same sampling method were repeated many times, approximately 95% of those intervals would contain the true population parameter. The margin of error, therefore, helps communicate how much uncertainty there is in the result.

5 : Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

The purpose of an ANOVA (Analysis of Variance) test is to determine whether there are statistically significant differences between the means of three or more independent groups. Unlike a t-test, which is limited to comparing two means, ANOVA extends hypothesis testing to situations with multiple groups, making it a powerful tool in experiments and research studies where several groups or treatments are compared simultaneously.​

For a valid ANOVA test, several assumptions must be met:


*   The data within each group should be normally distributed.

*   The variances across groups (homogeneity of variance) should be approximately equal.

*  Observations should be independent of each other, often ensured by good experimental design or random sampling.​

ANOVA tests the null hypothesis that all group means are equal. If the test finds significant evidence against this null hypothesis, it indicates that at least one group mean is different from the others, but it does not specify which groups differ. Post-hoc tests can be performed after ANOVA to identify the specific groups with differences.

6 : Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

Here is a Python program to perform a one-sample Z-test on a given dataset, with interpretation of the results:

```python
from math import sqrt
from scipy.stats import norm

# Sample data (replace this with your dataset)
sample_data = [12, 15, 14, 16, 13, 16, 15, 14, 13, 15]

# Hypothesized population mean
pop_mean = 14

# Population standard deviation (known)
pop_std = 2

# Sample statistics
sample_mean = sum(sample_data) / len(sample_data)
sample_size = len(sample_data)

# Calculate the Z statistic
z_stat = (sample_mean - pop_mean) / (pop_std / sqrt(sample_size))

# Calculate the p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is sufficient evidence that the sample mean differs from the population mean.")
else:
    print("Fail to reject the null hypothesis: There is insufficient evidence to say the sample mean differs from the population mean.")
```

Explanation:

*   Replace sample_data with your actual dataset.

*   pop_mean is the hypothesized mean under the null hypothesis.

*   pop_std is the known population standard deviation (required for Z-test).

*   The program calculates the Z-score and the two-tailed p-value.


*   If the p-value is less than the significance level (0.05), it rejects the null hypothesis, indicating the sample mean significantly differs from the population mean; otherwise, it fails to reject it.


This code provides a clear demonstration of performing a one-sample Z-test with Python.

 7 : Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.

Here is a Python example to simulate a dataset from a binomial distribution with parameters n=10 and p=0.5
using NumPy, and then plot its histogram using Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # number of trials
p = 0.5  # probability of success
size = 1000  # number of samples to generate

# Simulate binomial data
data = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black', alpha=0.7)
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

This code generates 1000 samples of the number of successes in 10 trials with a success probability of 0.5, then plots the frequency distribution as a histogram. The bin edges are set so each integer count of successes is correctly centered in the bins.

8 : Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

Here is an example Python code that generates multiple samples from a non-normal distribution (exponential distribution), calculates the sample means, and demonstrates the Central Limit Theorem (CLT) by plotting the distribution of sample means approaching a normal distribution:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
population_size = 100000
sample_size = 30        # size of each sample
num_samples = 1000      # number of samples to draw

# Generate a population from a non-normal distribution (exponential)
population = np.random.exponential(scale=1.0, size=population_size)

# Draw multiple samples and compute their means
sample_means = [np.mean(np.random.choice(population, sample_size, replace=False)) for _ in range(num_samples)]

# Plot original population distribution (non-normal)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='orange')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot distribution of sample means (should approach normal distribution)
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='blue')
plt.title('Sampling Distribution of the Sample Mean (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Explanation:

* The population is sampled from an exponential distribution, which is skewed and not normal.


* We draw many samples of size 30 and calculate their means.

* According to the Central Limit Theorem, the distribution of these sample means will tend to be normal, even though the population is not.

* The plots show the original skewed population and the approximately normal sampling distribution of the sample mean.

This code visually demonstrates the Central Limit Theorem in action.

 9 :  Write a Python function to calculate and visualize the confidence interval
for a sample mean.

Here is a Python function that calculates and visualizes the confidence interval for a sample mean using Matplotlib:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation

    # Calculate the margin of error using t-distribution
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    # Plotting
    plt.figure(figsize=(6, 4))
    plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', color='blue', capsize=10)
    plt.xlim(0, 2)
    plt.xticks([])
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Value")
    plt.grid(axis='y')
    plt.show()

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=5, size=30)
plot_confidence_interval(sample_data)
```

This function:


*   Takes a dataset and confidence level (default 95%).

*   Calculates the sample mean and uses the t-distribution for margin of error.

*  Prints the confidence interval.

*   Visualizes the confidence interval with an error bar around the sample mean.









 10 : Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

Here is a Python example to perform a Chi-square goodness-of-fit test comparing observed and expected frequencies, along with explanation of the result:

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example data)
observed = np.array([50, 30, 20])

# Expected frequencies under the null hypothesis
expected = np.array([40, 40, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution significantly differs from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")
```

Explanation:


*  The Chi-square goodness-of-fit test assesses whether the observed frequency distribution differs from the expected distribution.

*   If the p-value is less than the significance level (e.g., 0.05), it suggests a significant difference meaning the observed data do not fit the expected distribution well.

*   Otherwise, there is insufficient evidence to say the distributions differ, indicating a good fit.
