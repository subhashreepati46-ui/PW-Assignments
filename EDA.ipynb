 1: What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.
(Hint: Compare their scope, techniques, and applications for each.)


Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are interconnected but distinct fields with different scopes, techniques, and applications.​

Artificial Intelligence (AI)

Scope: AI is the broadest concept that aims to create intelligent machines capable of mimicking human intelligence and performing tasks that typically require human cognition, such as reasoning, decision-making, and problem-solving. It encompasses ML, DL, and other reasoning-based systems.​

Techniques: AI uses rule-based systems, expert systems, search algorithms, natural language processing, robotics, and includes ML and DL as subsets.​

Applications: Chatbots, virtual assistants, robotics, self-driving cars, game-playing agents, and intelligent recommendation systems.​

Machine Learning (ML)

Scope: ML is a subset of AI focused on enabling computers to learn patterns from data without explicit programming for every task. It emphasizes training models to make predictions and improve performance over time with more data.​

Techniques: ML employs statistical models including regression, classification, clustering, decision trees, random forests, and support vector machines.​

Applications: Email spam detection, fraud detection, predictive analytics, sales forecasting, and customer recommendation engines.​

Deep Learning (DL)

Scope: DL is a specialized subset of ML that uses multi-layered neural networks to simulate human-like learning and handle complex, unstructured data. It requires massive datasets and significant computational power.​

Techniques: DL uses Convolutional Neural Networks (CNNs) for image processing, Recurrent Neural Networks (RNNs) for sequential data, and transformer models for natural language tasks.​

Applications: Face recognition, speech-to-text conversion, natural language translation, medical image diagnosis, and autonomous vehicle technology.​

Data Science (DS)

Scope: DS is an interdisciplinary field focused on extracting knowledge and actionable insights from both structured and unstructured data. It combines statistics, programming, domain expertise, and often leverages ML and AI tools.​

Techniques: DS involves data collection, cleaning, exploratory analysis, statistical modeling, data visualization, ML algorithms, and big data processing tools.​

Applications: Business analytics, market predictions, customer segmentation, healthcare analytics, risk analysis, and strategic decision-making.

2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

Overfitting happens when a model learns the training data “too well,” including noise, so it performs very well on training data but poorly on new data.
Underfitting happens when a model is too simple to capture the underlying pattern in the data, so it performs poorly on both training and test data.

1 Overfitting

*  Characteristics: Low training error, high validation/test error;
usually caused by a model that is too complex for the amount of data (high variance, low bias).

*  Detection:

   * Training accuracy much higher than validation/test accuracy.   

   * Learning curves where training loss keeps decreasing but validation loss starts increasing after some point.
   

* Prevention:

    *  Use regularization (L1/Lasso, L2/Ridge, dropout in neural networks).

    *  Use cross-validation to tune model complexity and hyperparameters.

    *  Simplify the model (fewer features, shallower trees, fewer layers).

    *  Collect more training data or use data augmentation.

    *  Early stopping when validation loss stops improving.

2 Underfitting

*   Characteristics: High error on both training and validation/test data; model is too simple (high bias, low variance).

*   Detection:

    * Training and validation accuracies are both low and close to each other.  

    * Learning curves where both training and validation losses stay high and do not improve much.



* Prevention:

    *   Increase model complexity (more features, deeper trees, more layers/units).

    *   Reduce regularization strength if it is too high.

    *   Train longer or with better features/feature engineering.


3 Bias-variance tradeoff, cross-validation, and regularization.

* Bias-variance tradeoff:

   *   Underfitting corresponds to high bias (model too simple), overfitting corresponds to high variance (model too sensitive to data).

   *   Good generalization means finding a balance where both bias and variance are reasonably low.



*   Cross-validation:

    * Split data into multiple folds and evaluate the model on different validation sets.

    * Helps detect overfitting (performance drop from training to validation) and select model/hyperparameters that generalize well.

*   Regularization:

    * Adds a penalty term to the loss function to discourage overly complex models (large weights).

    * Proper regularization reduces variance (controls overfitting) while keeping enough flexibility to avoid underfitting when tuned correctly.






























3: How would you handle missing values in a dataset? Explain at least three
methods with examples.

Missing values can be handled in several ways depending on how much data is missing, the type of feature, and how important that feature is for the model. Three commonly used methods are deletion, simple imputation (mean/median/mode), and predictive-model-based imputation.​

1. Deletion (Dropping rows/columns)

* Idea: Remove records (rows) or variables (columns) that contain missing values.​

* Example: If 2% of rows have a missing “Age” value, you may drop those rows and train the model on the remaining 98%.

* When to use:

   * Missing percentage is small and appears random.

   * The feature with many missing values is not critical, so you can drop the entire column.

* Pros/cons: Simple and keeps only complete data, but can reduce sample size and introduce bias if data is not missing completely at random.​



2.Mean/Median/Mode Imputation

* Idea: Fill missing values with a summary statistic of that feature so you keep all rows.​

* Examples:

  * For numerical features like “Age”:

    * Mean imputation: replace missing ages with the average age.

    * Median imputation: better if the distribution is skewed (e.g., “Income”).

  *  For categorical features like “City”:

     * Mode imputation: replace missing entries with the most frequent city.

* When to use:

  * Missing percentage is moderate.

  * Relationships between features are not very complex or you are doing quick baselines/EDA.

* Pros/cons: Very easy and keeps dataset size, but it distorts variance and ignores relationships between features, which can slightly bias models.​

3. Predictive Modeling Imputation

* Idea: Treat the missing value as a prediction problem and use other features to estimate it.​

* Methods:

  * Regression imputation for numeric variables (e.g., predict “Age” using “Salary”, “Education”).​

  * Classification models for categorical variables (e.g., predict missing “Gender” label).

  * K-Nearest Neighbors (KNN) imputation: find k similar rows based on observed features and use their average/mode to fill in the missing value.​

* Example: To impute missing “Blood Pressure”, train a regression model on rows where it is known using features like “Age”, “Weight”, and “Cholesterol”, then use that model to predict the missing values.

* When to use:

  * When missingness is significant and feature relationships are important.

  * When you need higher-quality imputations for downstream ML models.

  * Pros/cons: Utilizes multivariate relationships and often gives more accurate imputations, but is more computationally expensive and requires careful validation.​

In practice, you first analyze the pattern and percentage of missing data, then choose the simplest appropriate method (often deletion or mean/median/mode) for minor gaps and switch to predictive modeling or more advanced methods when missingness is large or strongly related to other features.​

4: What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical). Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

An imbalanced dataset is a classification dataset where one class has many more examples than the other(s), so a model can get high accuracy by mostly predicting the majority class while performing poorly on the minority class (for example, 95% “no fraud” vs 5% “fraud”). This is common in fraud detection, medical diagnosis, and anomaly detection, and it requires special techniques to make the model pay attention to the minority class.​

Random Over/Undersampling

Theory:

 * Oversampling increases the number of minority-class samples (by duplicating or synthesizing them) so that class counts become more balanced.​

 * Undersampling reduces the number of majority-class samples by randomly removing some of them to match the minority class.​

Practical example:

 * Suppose you have 9000 “no fraud” and 1000 “fraud” transactions.

   * Random oversampling: randomly duplicate minority “fraud” records until you reach 9000 fraud and 9000 non-fraud, then train a classifier (e.g., logistic regression) on this balanced data.

   * Random undersampling: randomly drop majority “no fraud” examples down to 1000 so you train on 1000 fraud + 1000 non-fraud.

* Oversampling keeps all information but can overfit to duplicated samples; undersampling reduces training time but may discard useful majority information.​

SMOTE (Synthetic Minority Over-sampling Technique)

Theory:

* SMOTE is a smarter oversampling method that generates new synthetic minority samples instead of just copying existing ones.​

* For each minority example, SMOTE selects its nearest minority neighbors in feature space and creates synthetic points along the line segments between them, which makes the minority region denser and smoother.​

Practical example:

* In a credit-risk dataset, “default” might be the minority class. You first split data into train/test, then apply SMOTE only to the training set using a library function (e.g., SMOTE() in Python).

* After SMOTE, retrain a classifier such as Random Forest on the augmented training data and evaluate using metrics like precision, recall, F1-score, and ROC-AUC, which usually improve for the minority class compared to training on the original imbalanced data.​

Class Weights in Models

Theory:

* Instead of changing the data, you change the learning algorithm so that misclassifying minority samples is penalized more heavily than misclassifying majority samples.​

* Many models (logistic regression, SVM, tree-based models, neural networks) support a “class_weight” or cost-sensitive setting that increases the loss contribution from minority-class errors.​

Practical example:

* In scikit-learn, you can train LogisticRegression(class_weight="balanced") or provide custom weights such as {0:1, 1:5} where class 1 (minority) has 5× higher penalty.

* During training, the model adjusts its decision boundary to reduce minority misclassifications, often improving recall for the rare class without needing to resample the data.​

In practice, handling imbalanced data often combines these ideas: for example, applying SMOTE or oversampling on the training set together with appropriate class weights and evaluating using recall, precision, F1, and confusion matrices rather than accuracy alone.

Feature scaling is important because it puts all numeric features on a comparable scale, so no feature dominates just because it has larger units, and algorithms that use distances or gradient descent can learn faster and more accurately. It especially affects KNN, SVM, K‑means, neural networks, and any model trained with gradient descent.

* Distance‑based algorithms (KNN, SVM, K‑means): These use Euclidean or similar distances; if one feature ranges from 1–10,000 (salary) and another from 0–100 (age), the large‑range feature completely dominates the distance unless both are scaled.

* Gradient descent: When features have very different scales, the loss surface becomes stretched, causing very slow or unstable convergence; scaling makes gradients more balanced, so the algorithm reaches the optimum faster.

* When useful:

* For algorithms that expect features in a bounded range, such as neural networks with sigmoid/tanh activations, KNN, and distance‑based methods.

* When you want all inputs strictly in

Standardization (Z‑score scaling)

* Definition/formula: Transforms each feature to have mean 0 and standard deviation 1: x′=x−μσx ′ = σ x−μ.

* Properties:

   * Centers data around zero and rescales spread; works well if the feature is roughly Gaussian.

   * Less affected by single extreme values than Min‑Max, though still not completely robust to heavy outliers.

* When useful:

   * For linear models, SVM, logistic regression, PCA, and most gradient‑descent‑based methods, where having zero‑mean, unit‑variance features improves convergence and makes regularization (L1/L2) behave consistently across features.

 6:  Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

Label Encoding and One-Hot Encoding are two methods for converting categorical variables into numbers so machine learning models can use them, but they work differently and are best for different types of data.​

Label Encoding

* How it works: Assigns a unique integer to each category (e.g., "Red" → 0, "Green" → 1, "Blue" → 2).​

* Output: Single column with integer values replacing the original categories.​

* Pros:

  * Memory efficient; no extra columns created.

  * Fast and simple to implement.

* Cons:

  * Introduces an artificial ordinal relationship (the model may think 2 > 1 > 0, implying "Blue" > "Green" > "Red"), which can mislead algorithms that interpret numbers as having order or magnitude.

One-Hot Encoding

* How it works: Creates a new binary column for each category; each row gets a 1 in the column for its category and 0 in all others (e.g., "Red" →, "Green" →, "Blue" → ).

* Output: Multiple binary columns (one per category).​

* Pros:

  * No artificial ordering; each category is independent.

  * Works well with algorithms that treat features as continuous or calculate distances (e.g., linear regression, neural networks, KNN).

* Cons:

  * Increases dimensionality and memory usage, especially with high-cardinality features (many unique categories).

  * Can lead to sparse data (many zeros).

When to use each

* Use Label Encoding when the categorical variable is ordinal (has a natural order), e.g., Low < Medium < High, education levels.

* Use One‑Hot Encoding when the categorical variable is nominal (no natural order), e.g., colors, countries, product types.

* Use Label Encoding when using tree‑based models (Decision Trees, Random Forests, XGBoost) that can handle numeric labels well.

* Use One‑Hot Encoding when using distance‑based or linear models (KNN, SVM, logistic regression, neural networks), where numeric labels would wrongly imply order.

* Use Label Encoding when you have high‑cardinality features and need to keep the number of columns small.

* Use One‑Hot Encoding when you have few categories and want to avoid introducing any fake ordering between them.

Example:

* For a "Temperature" feature with values "Cold", "Warm", "Hot" (ordinal), use Label Encoding because Cold < Warm < Hot.

* For a "City" feature with values "Mumbai", "Delhi", "Bangalore" (nominal), use One-Hot Encoding because there is no natural ordering among cities and you don't want the model to assume Mumbai < Delhi.

b) Python code (Colab‑ready)
Paste this into a Colab cell and run it step by step. It clones the repo, reads the CSV, cleans ratings, and analyzes the relationship between category and rating.

```python
# Q7: Google Play Store – category vs rating

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# a) Load dataset from GitHub repo
!git clone https://github.com/MasteriNeuron/datasets.git
data_path = "/content/datasets/googleplaystore.csv"

df = pd.read_csv(data_path)

# Quick look
print(df.head())
print(df.info())

# b) Keep only valid numeric ratings
df = df.copy()
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df = df[df["Rating"].notna()]

# Drop rows where Category is missing
df = df[df["Category"].notna()]

# c) Group by Category and compute mean rating
cat_rating = (
    df.groupby("Category")["Rating"]
      .mean()
      .sort_values(ascending=False)
)

print("Average rating by category (descending):")
print(cat_rating)

# Top 5 highest and lowest categories
print("\nTop 5 highest rated categories:")
print(cat_rating.head())

print("\nTop 5 lowest rated categories:")
print(cat_rating.tail())

# d) Visualization – bar plot of average rating by category
plt.figure(figsize=(12, 6))
sns.barplot(
    x=cat_rating.index,
    y=cat_rating.values,
    color="skyblue"
)
plt.xticks(rotation=90)
plt.ylabel("Average Rating")
plt.title("Average App Rating by Category")
plt.tight_layout()
plt.show()
```

The goal is to use the Titanic CSV from that GitHub repo in Colab, compute survival rates by class and by age group (child vs adult), and then interpret the results. Typical Titanic analyses consistently show higher survival in higher classes and for children.​

a) Survival vs Passenger Class (Pclass)

* In the classic Titanic dataset, 1st class has the highest survival rate, followed by 2nd class, and 3rd class has the lowest survival rate (for example, around 60–65% in 1st, 40–45% in 2nd, and 20–25% in 3rd, depending on exact preprocessing).​

* Likely reasons:

   Code (Colab – part a)

```python
import pandas as pd

# Clone repo and load Titanic dataset
!git clone https://github.com/MasteriNeuron/datasets.git
titanic_path = "/content/datasets/titanic.csv"   # adjust name if different
df = pd.read_csv(titanic_path)

# Keep only needed columns
df = df[["Survived", "Pclass", "Age"]]

# a) Survival rate by passenger class
pclass_surv = (
    df.groupby("Pclass")["Survived"]
      .mean()
      .sort_index()
)

print("Survival rate by passenger class:")
print(pclass_surv)

# Nice percentage view
print("\nSurvival rate (%) by class:")
print((pclass_surv * 100).round(2))
```

b) Survival vs Age (Children vs Adults)

Most analyses of the Titanic dataset show that children survived at a higher rate than adults, consistent with the “women and children first” policy (e.g., about 55–60% of children versus 35–40% of adults, depending on age cut‑off).​

Code (Colab – part b)

```python
# b) Define age groups: children (<18) and adults (>=18)
# Drop rows with missing Age first
age_df = df.dropna(subset=["Age"]).copy()

age_df["AgeGroup"] = age_df["Age"].apply(
    lambda x: "Child" if x < 18 else "Adult"
)

# Survival rate by age group
age_surv = (
    age_df.groupby("AgeGroup")["Survived"]
          .mean()
)

print("\nSurvival rate by age group:")
print(age_surv)

print("\nSurvival rate (%) by age group:")
print((age_surv * 100).round(2))
```

a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git

Flight prices usually rise as the departure date gets closer, often with sharp increases in the last 2–3 weeks, and airlines differ in price levels depending on whether they are low‑cost or full‑service carriers. The given dataset lets you see this pattern numerically and compare airlines on a route like Delhi–Mumbai.​

a) Price vs days to departure
* Theory: Airline revenue management systems keep cheaper “fare buckets” open far from departure and progressively close them as seats fill, causing prices to increase, sometimes almost exponentially in the last 21, 14, and 7 days before departure.​

* Typical “best booking window”: Many empirical studies find that booking roughly 1–3 months before departure gives the lowest average fares for many routes; prices often surge in the last 2–3 weeks and on the final few days before flying.​

Example Colab code (adapt to the exact CSV name/columns in that repo):

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (assumes cloned from repo)
flight_path = "/content/datasets/flight_price.csv"
df = pd.read_csv(flight_path)

print("Dataset columns:", df.columns.tolist())
print("\nSample data:")
print(df.head())

# a) Price vs days_left (full analysis)
days_price = df.groupby('days_left')['price'].mean().reset_index()
days_price['days_left'] = pd.to_numeric(days_price['days_left'])

plt.figure(figsize=(10, 6))
sns.lineplot(data=days_price, x='days_left', y='price')
plt.gca().invert_xaxis()
plt.axvline(x=60, color='r', linestyle='--', label='Best window: ~60 days')
plt.xlabel('Days Left to Departure')
plt.ylabel('Average Price')
plt.title('Flight Price vs Days to Departure')
plt.legend()
plt.tight_layout()
plt.show()

print("Average prices by time periods:")
print(days_price[days_price['days_left'].isin([1, 7, 14, 30, 60, 90])])
```

* “Prices are relatively stable when there are more than about 30–45 days left, then start increasing as departure approaches, with a steep surge in the last 7–14 days. This suggests an exponential‑like increase close to departure, so the recommended booking window is around 1–2 months before flying to avoid last‑minute price spikes.”​

b) Prices across airlines for a fixed route (Delhi–Mumbai)

* Real‑world pattern: On routes like Delhi–Mumbai, low‑cost carriers (e.g., IndiGo, SpiceJet, Akasa, Air India Express) tend to offer lower base fares, while full‑service airlines (e.g., Air India, Vistara) generally charge more but provide extras like meals, baggage, and better service.​

* Reasons: Cost structure (no‑frills vs full service), brand positioning, included services, and loyalty programs all influence whether an airline is “budget” or “premium.”​

Example code to compare prices for Delhi–Mumbai:

```python
delhi_mumbai = df[
    (df['source_city'] == 'Delhi') &
    (df['destination_city'] == 'Mumbai')
].copy()

airline_prices = delhi_mumbai.groupby('airline')['price'].agg(['mean', 'count']).round(0)
airline_prices = airline_prices.sort_values('mean')

print("\nDelhi-Mumbai average prices by airline:")
print(airline_prices)

plt.figure(figsize=(10, 6))
sns.barplot(data=delhi_mumbai, x='airline', y='price')
plt.xticks(rotation=45)
plt.title('Average Flight Prices: Delhi to Mumbai by Airline')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()
```
